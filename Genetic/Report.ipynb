# What is G.A?
In this project we are trying to solve searching problems with natural selection way called Genetic Algorithm. This algorithm basically used in the problems where data and searching nodes are bigger than usual and it helps us to find the most optimal answers in almost short time. GA is based on natural selection. we have initial chromosomes(finite set of answers) which can be coorect or not. Then we try to combine these answers in order to approach to better ones until we find the optimal. we should have a "Fittnes function" to find out how good each answer is. Then we have to have a system to generate new chromosomes out of the last ones.

# Primitives
## Decleration of chromosome
Now we started to declare the variables of the problem. First of all, we have to represent our chromosomes. Each chromosome is an answer to the problem and the answer of the problem is 9 different gates that handle all of the outputs out of its inputs in dataframe. So our chromosome is an array of 9 digits between 0 and 5 where each of them represents one of the gates. To be clear we have this table:

| Digit | Gate |
| :-: | :-: |
| 0 | And |
| 1 | Nand |
| 2 | Or |
| 3 | Nor |
| 4 | Xor |
| 5 | Xnor |

In the first generation we can have any number of chromosomes  to start. The challange is to find out is it always better to have more chromosomes to start?

## Define the fittnes function
Next step is to declare the fittnes function. We have to assign a value to each chromosome to represent how close this answer to the optimal one is. In order to do so, we use the given dataframe. For each line we give the inputs to out answer and evaluate its output. If the output is correct one point will be added to answers value. If the value of an answer is equal to 1024 it is optimal. The better answer pass more logic calculations. Obviously for all cases better answers has greater value. Hence our function works good.

## Generating new chromosomes
In this stage we are about to evaluating next generation chromosomes from current ones. This is the main part of the project which make considerable effect on time and space complexity. Here we try two main ways to combine solutions.
Here we have our own decision to make. What we do in this stage is going to describe. In each state we have N chromosomes and we want to make N new ones. At first we sort  our solutions by their fittnes value. Then we devide them into 3 parts. The first part including better answers. We try to maintain these ones on next generation.So we copy the exact top part chromosomes into next geneartion. Then we have medium part. This solutions have the potential to become optimal. So we mutate them and add it to next generation. The worst part disappoint us so we completly forget them. There is still one third of next generation that is empty. Here is where we use crossover algorithm to fill empty places. This part of the next generation filled by combination of the answers of first part(which consist of best solution we have).

#### OnePoint Crossover 
Which randomly choose a location in the solution and devide new answer to two part.It picks one part of the answer from solution1 and the other one from solution2.
#### TwoPoint Crossover
Which is almost likely to previous one except there is two random location and new solution devide to 3 parts. Each of them taken from one solution.
#### Muatation
This method is used to reach randomly new answers. Whitout muatation it is possible for program to stick in same solutions and unable to reach better answers. We start this function with choosing one random location in chomosome to change the gene. Then we have to randomly change this gene's value. We can also change two or three genes in chromosome.

All of above algorithms were checked in the program and compare to each other in case of time duration and steps covered to reach optimal answer.

```python
import pandas as pd
import random
import time
```

```python
CHROMLEN = 9
CHROMNUM = 21
CHROMNUM2 = 9
## These are hyper values and can be change.
```

```python
def isCorrect(sample, solution):
    input1 = solution[0]
    j = 0
    for i in range(len(sample)):
        if sample[i]==0:
            j = int( (input1 + solution[i+1]) == 2 )
        if sample[i]==1:
            j = int( (input1 + solution[i+1]) != 2 )
        if sample[i]==2:
            j = int( (input1 + solution[i+1]) != 0 )
        if sample[i]==3:
            j = int( (input1 + solution[i+1]) == 0 )
        if sample[i]==4:
            j = (input1 + solution[i+1]) % 2
        if sample[i]==5:
            j = (input1 + solution[i+1] + 1) % 2
        ##print('operation =', sample[i],'input =', input1, 'input =', solution[i+1], 'output =', j)
        input1 = j
    if j==solution[-1]:
        return 1
    else:
        return 0
# Finding wheter the solution can pass the test or not
```

```python
def fitness(solution, df):
    s=0
    for i in df:
       s += isCorrect(solution, i)
    return s
# Calculating the value of the solution
```

```python
def mutate(l):
    p = random.randrange(0, CHROMLEN)
    l[p] = random.randrange(0, 6)
    return l
# Mutate function which change the value of one gene
```

```python
def mutate2(l):
    p = random.randrange(0, CHROMLEN)
    l[p] = random.randrange(0, 6)
    p = random.randrange(0, CHROMLEN)
    l[p] = random.randrange(0, 6)
    return l
# Mutate function which change the value of two gene2
```

```python
def onePointmix(l1, l2):
    point = random.randrange(0, CHROMLEN)
    out = []
    for i in range(len(l1)):
        if i <= point:
            out.append(l1[i])
        else:
            out.append(l2[i])
    return out
# performing onePoint crossover
```

```python
def twoPointmix(l1, l2):
    point = random.randrange(0, CHROMLEN)
    point2 = random.randrange(0, CHROMLEN)
    if point>point2:
        temp = point
        point = point2
        point2 = temp
    out = []
    for i in range(len(l1)):
        if i <= point:
            out.append(l1[i])
        elif i < point2:
            out.append(l2[i])
        else:
            out.append(l1[i])
    return out
# Performing twoPoint crossover
```

```python
def generateNextGen(current, val):
    newgen = []
    for i in range(0, 3):
        newgen.append(current[val[i]])
    for i in range(7, 9):
        k = mutate(current[val[i]])
        newgen.append(k)
    for i in range(9, 17):
        newgen.append(mutate(onePointmix(current[val[i-9]], current[val[i-8]])))
        newgen.append(mutate(onePointmix(current[val[i - 8]], current[val[i - 9]])))
    return newgen
# Generating next generation when each generation has 9 chromosomes
```

```python
def generateNextGen2(current, val):
    newgen = []
    for i in range(0, 3):
        newgen.append(current[val[i]])
    for i in range(7, 9):
        k = mutate(current[val[i]])
        newgen.append(k)
    for i in range(9, 17):
        newgen.append(mutate(twoPointmix(current[val[i-9]], current[val[i-8]])))
        newgen.append(mutate(twoPointmix(current[val[i - 8]], current[val[i - 9]])))
    return newgen
# Generating next generation when each generation has 21 chromosomes
```

```python
def generateNextGen3(current, val):
    newgen = []
    for i in range(0, 3):
        newgen.append(current[val[i]])
    for i in range(7, 9):
        k = mutate2(current[val[i]])
        newgen.append(k)
    for i in range(9, 17):
        newgen.append(mutate2(onePointmix(current[val[i-9]], current[val[i-8]])))
        newgen.append(mutate2(onePointmix(current[val[i - 8]], current[val[i - 9]])))
    return newgen
# Generating next generation when each generation has 21 chromosomes and using mutate2
```

```python
def generateNextGen4(current, val):
    newgen = []
    for i in range(0, 3):
        newgen.append(current[val[i]])
    for i in range(3, 6):
        k = mutate(current[val[i]])
        newgen.append(k)

    newgen.append(mutate(onePointmix(current[val[1]], current[val[0]])))
    newgen.append(mutate(onePointmix(current[val[0]], current[val[2]])))
    newgen.append(mutate(onePointmix(current[val[1]], current[val[2]])))
    return newgen
# Generating next generation when each generation has 9
```

```python
def generateNextGen5(current, val):
    newgen = []
    for i in range(0, 3):
        newgen.append(current[val[i]])
    for i in range(7, 9):
        k = mutate2(current[val[i]])
        newgen.append(k)
    for i in range(9, 17):
        newgen.append(mutate2(twoPointmix(current[val[i-9]], current[val[i-8]])))
        newgen.append(mutate2(twoPointmix(current[val[i - 8]], current[val[i - 9]])))
    return newgen
# Generating next generation when each generation has 21 chromosomes
```

```python
def printSol(l):
    for i in l:
        if i==0:
            print('And', end=' ,')
        if i==1:
            print('Nand', end=' ,')
        if i==2:
            print('Or', end=' ,')
        if i==3:
            print('Nor', end=' ,')
        if i==4:
            print('Xor', end=' ,')
        if i==5:
            print('Xnor', end=' ,')
# Encoding digits into the gates
```
